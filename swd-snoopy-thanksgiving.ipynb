```python
# Load packages
import pandas as pd
import numpy as np
import plotly
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
import matplotlib.colors
import seaborn as sns
import sklearn
from PIL import Image
from PIL import ImageOps
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
#Picture handling for snoopy
orig_img_file_path = r"C:\Users\nickp\OneDrive\Desktop\Home Data Projects\Viz\swdSnoopyThanksgiving\\"
orig_img_name = "nasa_snoopy_photo.png"
orig_img = Image.open(orig_img_file_path+orig_img_name)

thumbnail_name = "nasa_snoopy_thumbnail.png"
snoopy_size = 420,420
orig_img.thumbnail(snoopy_size)
orig_img.save(orig_img_file_path+thumbnail_name)
#orig_img.show()
```

```python
handler_img_path = r"C:\Users\nickp\OneDrive\Desktop\Home Data Projects\Viz\swdSnoopyThanksgiving\\"
handler_img_name = "handler.png"
handler_img = Image.open(handler_img_path+handler_img_name)

handler_thumbnail = "handler_thumbnail.png"
handler_size = 35,35
handler_img.thumbnail(handler_size)
handler_img.save(handler_img_path+handler_thumbnail)
#handler_img.show()
```

```python
# Make borders
snoopy_astronaut_height = 43
snoopy_astronaut_width = 29
snoopy_astronaut_handlers = 90
snoopy_handlers_ten = int(snoopy_astronaut_handlers / 10)
handler_spacer = snoopy_astronaut_width + 4
#print(snoopy_handlers_ten)
```

```python
fig, ax = plt.subplots()
plt.rcParams["figure.figsize"] = [14, 10]
plt.rcParams["font.size"] = 14
plt.rcParams["font.family"] = 'Arial'

# ax2 = ax.twinx()
# ax2.set_ylim(0,snoopy_handlers_ten+1)
def getImage(path):
   return OffsetImage(plt.imread(path, format="png"), zoom=1)

#add snoopy
snoop = AnnotationBbox(getImage(orig_img_file_path+thumbnail_name), (snoopy_astronaut_width/2 + 1
                                                                  ,snoopy_astronaut_height/2 + 2), frameon=False)
ax.add_artist(snoop)
#add handlers
for i in range(1,snoopy_handlers_ten+1):
    hndlr = AnnotationBbox(getImage(handler_img_path+handler_thumbnail)
                           ,((i)*(snoopy_astronaut_width/snoopy_handlers_ten)-.5,3)
                           ,frameon=False)
    ax.add_artist(hndlr)
#add height and width lines
plt.vlines(x=0,ymin=0,ymax=snoopy_astronaut_height,colors='#BC1F1F',linewidth=20)
plt.hlines(xmax=snoopy_astronaut_width,xmin=0,y=0,colors='#804000',linewidth=20)

#labels
plt.text(0,snoopy_astronaut_height+2,str(snoopy_astronaut_height)+" ft Tall"
         ,horizontalalignment='center',verticalalignment='center',fontsize=14, c='black',weight='bold')

plt.text(snoopy_astronaut_width/2,-2,str(snoopy_astronaut_width)+" ft Wide"
         ,horizontalalignment='center',verticalalignment='center',fontsize=14, c='black',weight='bold')

plt.text(snoopy_astronaut_width+1,5,str(snoopy_astronaut_handlers)+" float handlers"
         ,horizontalalignment='left',verticalalignment='center',fontsize=14, c='black')

plt.axis('off') 
plt.ylim(0,50)
plt.xlim(0,50)

plt.title(label="Saying Goodbye to Astro Snoopy (with matplotlib)!",
         loc="left",
         fontsize=18,
         weight="bold")

plt.show
```
